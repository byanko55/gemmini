### Test1: Spherical

```python
import gemmini as gm
from math import pi, cos, sin

fa = gm.Circle(r=5)
fb = gm.Arc(r=4, angle=3*pi/4, n=16)
fc = gm.Ellipse(size=(4, 8), n=48)
fd = gm.Ellipse(height=8, width=4, n=48)

for f in [fa, fb, fc, fd]:
    canva = gm.Canvas()
    canva.add(f)
    canva.plot()
```

### Test2: Spiral

```python
fa = gm.Spiral(10, 6*pi, 150)
fb = gm.ParabolicSpiral(10, 8*pi, 99)
fc = gm.ParabolicSpiral(10, 8*pi, 100)
fd = gm.LituusSpiral(10, 5*pi, 150)

for f in [fa, fb, fc, fd]:
    canva = gm.Canvas()
    canva.add(f)
    canva.plot()
```

```python
fa = gm.HyperbolicSpiral(10, 4*pi, n=64)

canva = gm.Canvas()
canva.add(fa)
canva.plot()
```

```python
fa = gm.LogarithmicSpiral(10, 2*pi)

canva = gm.Canvas()
canva.add(fa)
canva.plot()
```

```python
f = gm.BoundedSpiral(10, 8*pi, 256)

canva = gm.Canvas()
canva.add(f)
canva.plot()
```

### Test3: Cycloid

```python
f = gm.Cycloid(5, 6*pi, n=128)

canva = gm.Canvas()
canva.add(f)
canva.plot()
```

```python
fa = gm.Epicycloid(p=5, q=3, size=16)
fb = gm.Epicycloid(p=2, q=1, size=16)
fc = gm.Hypocycloid(p=5, q=3, size=16)
fd = gm.Hypocycloid(p=7, q=4, size=16)

for f in [fa, fb, fc, fd]:
    canva = gm.Canvas()
    canva.add(f)
    canva.plot()
```
