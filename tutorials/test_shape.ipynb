### Test1: Circular Sector/Segment

```python
import gemmini as gm
from math import pi

fa = gm.CircularSector(r=5, a=4*pi/3)
fb = gm.CircularSegment(r=5, a=2*pi/3)
fc = gm.CircularSegment(r=5, a=pi)
fd = gm.CircularSegment(r=5, a=3*pi/2)

for f in [fa, fb, fc, fd]:
    canva = gm.Canvas()
    canva.add(f)
    canva.plot()
```
