```python
import gemmini as gm
from math import pi

def get_all_gem():
    fig = []

    fig.append(gm.RegularPolygon(4, num_vertex=5))
    fig.append(gm.RegularPolygon(4, num_vertex=6))
    fig.append(gm.RegularPolygon(3, num_vertex=8))

    fig.append(gm.Parallelogram(s=5, n=48))
    fig.append(gm.Rhombus(h=6, w=10))

    fig.append(gm.IsoscelesTriangle(size=(7, 5)))
    fig.append(gm.RightTriangle(size=(5, 7)))

    fig.append(gm.Trapezoid(s=(4, 6, 7, 6), n=36))
    fig.append(gm.Rectangle(h=7, w=6, n=36))

    fig.append(gm.Kite(s=(8, 7), n=36))
    fig.append(gm.ConcaveKite(h=8, w=7, n=36))
    fig.append(gm.ConcaveStar(s=5, num_vertex=5, num_dot=7))

    fig.append(gm.Circle(r=4, n=48))
    fig.append(gm.Ellipse(size=(4, 8)))

    fig.append(gm.Spiral(4, 6*pi, 100))
    fig.append(gm.ParabolicSpiral(5, 8*pi, 100))

    fig.append(gm.BoundedSpiral(4, 8*pi, 100))

    fig.append(gm.Epicycloid(p=5, q=3, size=9))
    fig.append(gm.Epicycloid(p=2, q=1, size=9))
    fig.append(gm.Hypocycloid(p=5, q=3, size=9))

    fig.append(gm.Hypocycloid(p=7, q=4, size=8))
    fig.append(gm.CurvedPolygon(size=8, num_vertex=5))
    fig.append(gm.Lissajous(a=3, b=2, size=8, n=256))

    fig.append(gm.Folium(r=8, n=128))
    fig.append(gm.Bifolium(r=10, n=96))

    fig.append(gm.CircularSector(r=4, a=4*pi/3, n=48))
    fig.append(gm.CircularSegment(r=4, a=3*pi/2, n=48))
    fig.append(gm.Wave(a=2, w=8, f=2, n=64))

    fig.append(gm.Helix(r=2.5, pitch=0.75, a=6*pi, n=128))
    fig.append(gm.SymmetricSpiral(r=10))

    fig.append(gm.Star(s=4))
    fig.append(gm.Star(s=4, v=6))
    fig.append(gm.Star(s=4, v=8))

    fig.append(gm.Heart(s=8))
    fig.append(gm.ButterFly(s=6))
    fig.append(gm.CottonCandy(s=8, c=6))

    fig.append(gm.Boomerang(s=8))
    fig.append(gm.Stellate(s=8))
    fig.append(gm.Stellate(s=8, c=16))

    fig.append(gm.Shuriken(s=8))
    fig.append(gm.Flower_A(s=8))
    fig.append(gm.Flower_B(s=8))
    fig.append(gm.Flower_C(s=8))
    fig.append(gm.Flower_D(s=8))
    fig.append(gm.Flower_E(s=8))
    fig.append(gm.Flower_F(s=8, n=256))

    fig.append(gm.FattyStar(s=6))
    fig.append(gm.Yinyang(s=8))
    fig.append(gm.Moon(r=8))
    fig.append(gm.Polygontile(s=4, v=4))
    fig.append(gm.Polygontile(s=4, v=5))
    fig.append(gm.Polygontile(s=4, v=6))

    fig.append(gm.Gear(r=4))
    fig.append(gm.SnippedRect(h=5, w=8, nD=128, clip_size=(1.0, 0, 0, 0)))
    fig.append(gm.RoundedRect(h=5, w=8, nD=256, border_radius=(0.6, 0.5, 0.8)))

    fig.append(gm.Plaque(s=8, nD=256))
    fig.append(gm.BlockArc(R=4, r=2, a=3*pi/2, n=100))

    fig.append(gm.Cross_A(s=8, w=2))
    fig.append(gm.Cross_B(s=8))
    fig.append(gm.Cross_C(s=8, n=20))
    fig.append(gm.SunCross(s=8))
    fig.append(gm.BasqueCross(s=8, n=64))
    fig.append(gm.CelticCross(s=8))

    fig.append(gm.Lshape(w=3))
    fig.append(gm.HalfFrame(h=5, w=8, breadth=1.5))
    fig.append(gm.Teardrop(s=6, n=100))
    fig.append(gm.Nosign(s=8, n=100))

    fig.append(gm.Arrow(h=3, w=10, n=100))
    fig.append(gm.DoubleArrow(h=3, w=10, n=100))
    fig.append(gm.ArrowPentagon(h=3, w=10, n=100))
    fig.append(gm.ArrowChevron((5, 5)))

    fig.append(gm.Clover(s=10, num_leaves=3))

    return fig
```

```python
canva = gm.Canvas(theme='solarized', draw_grid=False, window_size=(16, 16), scale=0.8)
fig = get_all_gem()

for i, f in enumerate(fig):
    f.translate(10*(i//6), 10*(i%6) + 5*((i//6)%2))

    if isinstance(f, gm.Yinyang):
        subfigs = f.interior()

        for _f in subfigs:
            canva.add(_f, fill=True, tickness=5)
    else :
        canva.add(f, fill=True, tickness=5)
    
canva.plot()
```
