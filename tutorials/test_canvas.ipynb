### Test1: Add/Remove geometry to canvas  

```python
import gemmini as gm
from math import pi

canva = gm.Canvas()
    
a = gm.Line2D((10,0), (0,15))
b = gm.Grid(10, 8, num_dot=(10, 8))
c = gm.Point2D(px=-6, py=9)

canva.add((a,b,c))
canva.plot()
```

```python
canva.remove(a)
canva.plot()
```

### Test2: Resizing (Zoom in/out) Window

```python
canva = gm.Canvas(scale=1.25)

canva.add((a,b,c))
canva.plot()
```

### Test3: Remove minor ticks

```python
canva = gm.Canvas(minor_ticks=False)
    
a = gm.Line2D((10,0), (0,15))
b = gm.Line2D((-1,-1), (5,2))
c = gm.RegularPolygon(6, 10, v=5)
c.translate(4, 8)

canva.add(a, color='skyblue')
canva.add(b)
canva.add(c)
canva.plot()
```

### Test4: Change Pixel/Line style

```python
canva = gm.Canvas()
    
a = gm.Line2D((10,0), (0,15))
b = gm.Line2D((-1,-1), (5,2))
c = gm.RegularPolygon(6, 10, v=5)
c.translate(4, 8)

canva.add(a, color='skyblue')
canva.add(b, style='--')
canva.add(c, style='*')
canva.plot()
```

### Test5: Hide Grid

```python
canva = gm.Canvas(draw_grid=False)
    
a = gm.Circle(r=3, nD=64)
b = gm.Ellipse(4, 8, 48)
a.translate(-6, -4)

canva.add((a, b), show_radius=True)
canva.plot()
```

### Test6: Apply Canvas theme

```python
canva = gm.Canvas(theme='gruvbox')
    
a = gm.RegularPolygon(s=3, v=7, n=3)
b = gm.RegularPolygon(s=3, v=6)
c = gm.RegularPolygon(s=3, v=5)

a.translate(0, 3)
b.translate(-8, 6)
c.translate(8, 0)

canva.add(a, show_edges=True)
canva.add(b, draw_interior=True)
canva.add(c, show_size=True)
canva.plot()
```

### Test7: Drawing Options

```python
f = gm.SunCross(s=9)
gm.plot((f), fill=True)

f.shatter((2, 4), rate=2)
gm.plot((f), fill=True)
gm.plot((f), show_area=True)
```

```python
canva = gm.Canvas(theme='horizon')

original = gm.IsoscelesTriangle(h=4, w=6)
figs = []

for i in range(6):
    f = original.copy()
    f.translate(8*(i%3), 8*(i//3))
    figs.append(f)

canva.add(figs[0], fill=True)
canva.add(figs[1], show_edges=True)
canva.add(figs[2], show_center=True)
canva.add(figs[3], show_radius=True)
canva.add(figs[4], show_size=True)
canva.add(figs[5], show_area=True)

canva.plot()
```
