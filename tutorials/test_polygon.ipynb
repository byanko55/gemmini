### Test1: Create a Polygon / Get Area

```python
import gemmini as gm
import numpy as np
from math import pi

f = gm.Polygon2D([[0,0], [3,0], [3,3], [2,1.5]])

assert f.area() == 3.75

canva = gm.Canvas()
canva.add(f, show_area=True)
canva.plot()
```

### Test2: Regular Polygon

```python
fa = gm.RegularPolygon(size=6, num_vertex=5)
fb = gm.RegularPolygon(size=6, num_vertex=6)
fc = gm.RegularPolygon(size=6, num_vertex=8)

for f in [fa, fb, fc]:
    canva = gm.Canvas()
    canva.add(f)
    canva.plot()
```

### Test3: Triangle

```python
fa = gm.RegularPolygon(6, num_vertex=3)
fb = gm.IsoscelesTriangle(size=(7, 5))
fc = gm.RightTriangle(size=(5, 9))

for f in [fa, fb, fc]:
    canva = gm.Canvas()
    canva.add(f)
    canva.plot()
```

### Test4: Basic Polygon

```python
fa = gm.Parallelogram(size=(9, 6), n=[5, 10, 5, 10], angle=pi/4)
fb = gm.Rhombus(h=6, w=10)
fc = gm.Rhombus(s=8, num_dot=[6,12,12,6])

for f in [fa, fb, fc]:
    canva = gm.Canvas()
    canva.add(f)
    canva.plot()
```

### Test5: Trapezoid

```python
fa = gm.Trapezoid(s=(8, 6))
fb = gm.Trapezoid(s=(8, 6, 10))
fc = gm.Trapezoid(s=(6, 8, 10, 12))
fd = gm.RightTrapezoid(s=(8, 6, 10))

for f in [fa, fb, fc, fd]:
    canva = gm.Canvas()
    canva.add(f)
    canva.plot()
```

### Test6: Rectangle

```python
fa = gm.Rectangle(h=8, w=6, n=40)
fb = gm.Rectangle(s=(8,6), n=[10, 8, 12, 6])

for f in [fa, fb]:
    canva = gm.Canvas()
    canva.add(f)
    canva.plot()
```
