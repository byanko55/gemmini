### Test1: Point/PointSet

```python
import gemmini as gm
import numpy as np

canva = gm.Canvas()
p = gm.Point2D(3,4)

points = 5*np.random.rand(20, 2)
pc = gm.PointSet2D(points)

print("bounding box: ", pc.bounding_box())
print("centroid: ", pc.center())
print("width x height:", pc.dim())
print("farthest distance from center (`radius` for spherical): ", pc.rad())
print("2~5th dots: ", pc[1:5])

canva.add(gem=(p, pc))
canva.plot()
```

### Test2: Pointcloud

```python
canva = gm.Canvas()

cluster_small = gm.Pointcloud2D(s=0.05, num_dot=10)
cluster_medium = gm.Pointcloud2D(s=0.1, num_dot=20)
cluster_large = gm.Pointcloud2D(s=0.2, num_dot=50)

canva.add((cluster_small, cluster_medium, cluster_large), show_center=True)
canva.plot()
```

### Test3: Grid

```python
canva = gm.Canvas()
g = gm.Grid((5, 4), num_dot=(10, 8))

canva.add(g)
canva.plot()
```
